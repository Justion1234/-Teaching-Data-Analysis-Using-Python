## 반복문



1.   While
2.   For



While문 실행

```python
count = 0
while count < 5:
  print("%d번째"%(count))
  count += 1
```

For문 실행

range()를 이용하는 반복문

```python
for i in range(0, 5):
  print('%d번째'%(i))
```

문자열을 이용하는 반복문

```python
for i in 'hello world':
  print(i)
```

리스트를 이용하는 반복문

```python
for i in [11, 22, 33, 44, 55]:
  print(i)
```

```python
var = {'key1': 'value1', 'key2': 'value2', 'key3': 'value3'}
for key in var:
  print(key)
  print(var[key])
```

```python
string_1 = "drink"
string_2 = "bar"

print("Let's go %s. Tell me when you find a %s" % (string_1, string_2))
```
