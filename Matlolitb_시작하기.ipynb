## Matplotlib 범례 표시하기

범례(Legend)는 그래프 데이터의 종류를 표시하기 위한 텍스트이다.


matplotlib.pyplot 모듈의 legend()함수를 사용해서 그래프에 범례를 표시할 수 있습니다.

### 기본사용

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend() #이 코드를 통해서 범례를 그래프에 나타내줌

plt.show()


```

### 위치 지정하기

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(1.0, 1.0)) #이 코드를 통해서 범례를 그래프에 나타내줌

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc=(0, 0)) #이 코드를 통해서 범례를 그래프에 나타내줌

plt.show()
```

loc를 튜플 형식으로 위치를 나타내주는데 문자열로도 위치 나타낼 수 있습니다.

## Matplotlib 그래프 영역 채우기

그래프의 영역을 채우는 세가지 함수



*   fill_between(): 두 수평 방향의 곡선 사이를 채움
*   fill_betweenx(): 두 수직 방향의 곡선 사이를 채움
*   fill(): 닥각형 영역을 채움


### 기본 사용 - fill_between()

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]


plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)


plt.show()
```

### 기본 사용 - fill_betweenx()

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 10]


plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)


plt.show()
```

### 두 그래프 사이 영역 채우기

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```

## 그리드 설정하기

데이터의 위치를 더 명확하게 나타내기 위해 그래프에 그리드(Grid,격자)를 표시할 수 있습니다.  
그냥 그리드를 쉽게 설명하자면 그래프를 그릴 때 보기 쉽게 눈금 있는 거라고 생각하시면 됩니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)


plt.plot(x, x, 'bo')
plt.plot(x, x**2, color="#e35f62", marker="*", linewidth=2)
plt.plot(x, x**3, color="springgreen", marker="^", markersize=9)
plt.grid(True)

plt.show()
```

## Matplotlib 눈금 표시하기

틱(Tick)은 그래프에 간격을 구분하기 위해 표시하는 눈금입니다.  
matplotlib.pyplot 모듈의 xticks(), yticks(), tick_params()함수를 이용해서 그래프에 눈금을 표시할 수 있습니다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)


plt.plot(x, x, 'bo')
plt.plot(x, x**2, color="#e35f62", marker="*", linewidth=2)
plt.plot(x, x**3, color="springgreen", marker="^", markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', ''])
#labels라는 파라미터를 이용해서 눈금 레이블을 문자열의 형태로 지정할 수 있습니다.
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```

## Matplotlib 수평선 수직선 그리기

수직선/수평선 그리는 함수들



*   axhline(): 축을 따라 수평선을 표시합니다.
*   axvline(): 축을 따라 수직선을 표시합니다.
*   hlines():  지정한 점을 따라 수평선을 표시합니다.
*   vlines():  지정한 점을 다라 수직선을 표시합니다.



### 수평선 그리기 - axhline(), hilnes()

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2, 'g--')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)


plt.show()
```
